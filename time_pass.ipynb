```python
import matplotlib.pyplot as plt

# Step 2: Define the data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Step 3: Create the plot
plt.plot(x, y, marker='o')

# Step 4: Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Step 5: Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 2: Define the data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]
y3 = [2, 2, 3, 3, 5]

# Step 3: Create the plot with multiple lines
plt.plot(x, y1, marker='o', linestyle='-', color='b', label='Line 1')
plt.plot(x, y2, marker='s', linestyle='--', color='r', label='Line 2')
plt.plot(x, y3, marker='^', linestyle='-.', color='g', label='Line 3')

# Step 4: Customize the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Complex Line Plot')
plt.grid(True)

# Step 5: Add a legend
plt.legend()

# Step 6: Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Define the data
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2*pi
y_sin = np.sin(x)

# Step 3: Create the plot with the sine function
plt.plot(x, y_sin, marker='o', linestyle='-', color='b', label='sin(x)')

# Step 4: Customize the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Function Plot')
plt.grid(True)

# Step 5: Add a legend
plt.legend()

# Step 6: Display the plot
plt.show()
```
